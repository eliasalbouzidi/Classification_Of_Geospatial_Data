Define here the coefficients for each model

```python
coefs = [4, 1, 5, 3, 4]
# CatBoost | LightGBM | XGBoost13 | XGBoost17 | XGBoost15

vote = False 
# True => use class votes. False => Use weighted sum of probabilities
```

______________________

```python

name = f"ensemble_{coefs[0]}_{coefs[1]}_{coefs[2]}_{coefs[3]}_{coefs[4]}"
if vote:
    name = f"ensembleVOTE_{coefs[0]}_{coefs[1]}_{coefs[2]}_{coefs[3]}_{coefs[4]}"
```

```python
import numpy as np
import pandas as pd

pd.set_option('display.max_columns', None)  
pd.set_option('display.max_rows', 100)  
```

```python
prob_catboost = np.load("data/PROBS_CATBOOST.npy")
prob_lightgbm = np.load("data/PROBS_LIGHTGBM.npy")
prob_xgboost_13 = np.load("data/PROBS_XGBOOST_13.npy")
prob_xgboost_17 = np.load("data/PROBS_XGBOOST_17.npy")
prob_xgboost_15 = pd.read_csv("data/PROBS_XGBOOST_15.csv")
```

```python
prob_xgboost_15.drop(columns=["Id"], inplace=True)
```

```python
prob_xgboost_15.to_numpy().argmax(axis=1)
```

```python
csv1NN = pd.read_csv("data/1-NN.csv")
```

```python
prob_catboost[:,:5].argmax(axis=1)
```

_______________________
Define here the parameters

```python
dfALL = pd.DataFrame()
dfALL["catboost"] = prob_catboost.argmax(axis=1)
dfALL["lightgbm"] = prob_lightgbm.argmax(axis=1)
dfALL["xgboost_13"] = prob_xgboost_13.argmax(axis=1)
dfALL["xgboost_17"] = prob_xgboost_17.argmax(axis=1)
dfALL["xgboost_15"] = prob_xgboost_15.to_numpy().argmax(axis=1)
dfALL["1NN"] = csv1NN["change_type"]

dfALL[name] = (coefs[0]*prob_catboost[:,:5] + coefs[1]*prob_lightgbm[:,:5] + coefs[2]*prob_xgboost_13[:,:5] + coefs[3]*prob_xgboost_17[:,:5] + coefs[4]*prob_xgboost_15.to_numpy()[:,:5]).argmax(axis=1)
if vote:
    weights = {'catboost': coefs[0], 'lightgbm': coefs[1], 'xgboost_13': coefs[2],'xgboost_17': coefs[3],'xgboost_15': coefs[4]}
    cols = []
    for col in weights:
        for it in range(weights[col]):
            cols.append(dfALL[col])
    weighted_df = pd.concat(cols, axis=1)
    dfALL[name] = weighted_df.mode(axis=1)[0].astype(int)

dfALL.loc[dfALL["1NN"] != 10, name] = dfALL.loc[dfALL["1NN"] != 10, "1NN"]

for i in range(5):
    dfALL[f"p_catboost_{i}"] = prob_catboost[:,i]
for i in range(5):
    dfALL[f"p_lightgbm_{i}"] = prob_lightgbm[:,i]
for i in range(5):
    dfALL[f"p_xgboost_13_{i}"] = prob_xgboost_13[:,i]
for i in range(5):
    dfALL[f"p_xgboost_17_{i}"] = prob_xgboost_17[:,i]
for i in range(5):
    dfALL[f"p_xgboost_15_{i}"] = prob_xgboost_15.to_numpy()[:,i]
```

```python
result = dfALL[name].to_numpy()
```

```python
pred_df = pd.DataFrame(result, columns=['change_type'])
pred_df.to_csv(f"submissions/{name}.csv", index=True, index_label='Id')
```
